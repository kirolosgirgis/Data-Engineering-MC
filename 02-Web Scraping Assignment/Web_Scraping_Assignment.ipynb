```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
base_url = 'https://books.toscrape.com/catalogue/page-{}.html'
```

```python
books_info = []

for n in range(1,51):
    scrape_url = base_url.format(n)
    res = requests.get(scrape_url)
    soup = BeautifulSoup(res.text, 'lxml')
    products = soup.select('.product_pod')
    
    for x in products:
        url = x.select('a')[1]['href']
        book_url = 'https://books.toscrape.com/catalogue/' + url
        book_res = requests.get(book_url)
        book_soup = BeautifulSoup(book_res.text, 'lxml')
        
        dic = {}
        rating_dic = {'One':1,'Two':2, 'Three':3, 'Four':4, 'Five':5}
        
        dic['book_title'] = book_soup.select('h1')[0].text
        dic['book_image'] = 'https://books.toscrape.com/' + book_soup.select('img')[0]['src']
        dic['star_rating'] = rating_dic[str(book_soup.select('.star-rating')[0]).split('\n')[0].split()[2][0:-2]]        
        dic['upc'] = book_soup.select('td')[0].text
        dic['price'] = float(book_soup.select('td')[3].text[2:])
        dic['availability'] = int(book_soup.select('td')[5].text.split()[2][1:])
        books_info.append(dic)
        
# books_info
```

```python
books_df = pd.DataFrame(books_info)
books_df.index += 1
books_df
```
